## 1.Series
- Series是一种类似与一维数组的对象，由下面两部分组成：
    - values:一组数组（ndarray类型）
    - index:相关的数据索引标签
    
### Series的创建
- 两种创建方式：
- （1）由列表或numpy数组创建
    - 默认索引为0到N-1的整数类型索引

```python
s = pd.Series([x for x in range(1,5)])
s
```

```python
# 通过index指定索引
s.index = list('abcd')
s
```

```python
s1 = pd.Series(['zhangsan','lisi','wangwu','zhaoliu'],index=list('abcd'))
s1
```

```python
#Series只能存放一维数据
s2 = pd.Series(np.random.randint(0,10,size=5),index=list('abcde'))
s2
```
