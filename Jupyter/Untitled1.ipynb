```python
#折线图、散点图
import matplotlib.pylab as pyl
import pandas as pd
x = [1,2,3,4,7,9]
y = [3,5,4,7,9,1]
#pyl.plot(x,y)#plot（x轴数据，y轴数据）折线图
#线条样式
'''
- 直线
-- 虚线
-. -. 形式
： 细小虚线
'''
x1 = [1,3,5,10,12,19]
y1 = [1,6,8,10,19,20]
#加标题 pyl.title('show')
#x轴名称pyl.xlabel('age')
#y轴名称pyl.xlabel('temp')
pyl.plot(x,y,'r:')
pyl.plot(x1,y1,'g')
pyl.title("show")
pyl.xlabel('age')
pyl.ylabel('temp')
#pyl.xlim(0,9)#x轴范围
#pyl.ylim(1,10)#y轴范围
pyl.show()
```

```python
#pyl.plot(x,y,'o')#散点图
#修改点的颜色
'''
c-cyan--青色
r-red--红色
m-magente-品红
g-green--绿色
b-blue--蓝色
y-yellow--黄色
k-black--黑色
w-white--白色
'''
#pyl.plot(x,y,'go')
#点的样式
'''
s--方形
h--六角形
H--六角形
*--星形
+--加号
x--x形
d--菱形
D--菱形
p--五角形
'''
pyl.plot(x,y,'g*')
pyl.show()
```
